```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#lets generate age and salary randomly and make one by stacking
```

```python
#generate age
age=np.random.randint(20,70,50)
age
```

```python
salary=np.random.randint(20000,70000,50)
salary
```

```python
#stacking them up to make one dataset
data=np.stack((age,salary),axis=1)
data
```

```python
#concat garya vayae duitai lai yeuati ma hunthyo
con=np.concatenate((age,salary))
con
```

```python
#lets plot histogram of the datas
#histogram of age before scaling

plt.hist(age,bins=10,edgecolor='black')
plt.show()
```

```python
#now lets do standarization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(data)

```

#Outliers in standarization

yeuta duita outliers banama aani tellai last ma haldim

```python
outliers=np.array([[100,1000000],[110,1500000]])
outliers_data=np.append(data,outliers,axis=0)
outliers_data
```

```python
scaler = StandardScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(outliers_data)
```

```python
scaled_ouliers_data=scaler.transform(outliers_data)
scaled_ouliers_data
```

```python
plt.subplot(1,2,1)
plt.scatter(outliers_data[:,0],outliers_data[:,1])
plt.title('before scaling')
plt.subplot(1,2,2)
plt.scatter(scaled_ouliers_data[:,0],scaled_ouliers_data[:,1])
plt.title('after scaling')
plt.show()
```

#Normalization
goal is same
to bring dataset to common scales whithout distorting the difference in ranges or losing information

normalization ma chai

    (x-xmin)/(xmax-xmin)

here all the data definitely lies between 0 and 1
so it is like squishing data into a rectangle,cube ,hypercube according to the dimension



```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(outliers_data)

# transform train and test sets
normalized_outliers_data = scaler.transform(outliers_data)
```

```python
normalized_outliers_data
```

```python
plt.subplot(1,2,1)
plt.scatter(outliers_data[:,0],outliers_data[:,1])
plt.title('before normalization')
plt.subplot(1,2,2)
plt.scatter(normalized_outliers_data[:,0],normalized_outliers_data[:,1])
plt.title('after normalization')
plt.show()
```
